```python
from mmengine.config import Config

cfg = Config.fromfile("mmdetection/configs/ssd/ssd512_coco.py")
```

```python
cfg.model.bbox_head.num_classes = 1

cfg.data_root = "mmdetection/data/boxes/"

metainfo = dict(
    classes = ('Box',),
    palette = [
        (220, 20, 60),
    ]
)

cfg.train_dataloader.dataset = dict(
        type=cfg.dataset_type,
        data_root=cfg.data_root,
        metainfo=metainfo,
        ann_file='train.json',
        data_prefix=dict(img='train/'),
        filter_cfg=dict(filter_empty_gt=True, min_size=32),
        pipeline=cfg.train_pipeline,
        backend_args=cfg.backend_args)

cfg.val_dataloader.dataset = dict(
        type=cfg.dataset_type,
        data_root=cfg.data_root,
        metainfo=metainfo,
        ann_file='val.json',
        data_prefix=dict(img='valid/'),
        test_mode=True,
        pipeline=cfg.test_pipeline,
        backend_args=cfg.backend_args)

cfg.val_evaluator = dict(
    type='CocoMetric',
    ann_file=cfg.data_root + 'val.json',
    metric='bbox',
    format_only=False,
    backend_args=cfg.backend_args)

# cfg.model.bbox_head.loss_cls=dict(
#             type='FocalLoss',
#             use_sigmoid=True,
#             gamma=2.0,
#             alpha=0.25,
#             loss_weight=1.0),
# cfg.model.bbox_head.loss_bbox=dict(type='L1Loss', loss_weight=1.0)

cfg.test_dataloader.dataset = dict(
        type=cfg.dataset_type,
        data_root=cfg.data_root,
        metainfo=metainfo,
        ann_file='test.json',
        data_prefix=dict(img='test/'),
        test_mode=True,
        pipeline=cfg.test_pipeline,
        backend_args=cfg.backend_args)

cfg.test_evaluator = dict(
    type='CocoMetric',
    ann_file=cfg.data_root + 'test.json',
    metric='bbox',
    format_only=False,
    backend_args=cfg.backend_args)

cfg.work_dir = "work_dirs/ssd512_1xb2_24e_boxes"

cfg.train_cfg = dict(type='EpochBasedTrainLoop', max_epochs=24, val_interval=1)
cfg.auto_scale_lr = dict(enable=True, base_batch_size=8)
```

```python
from mmengine.runner import Runner
model = Runner.from_cfg(cfg)
model.train()
```
