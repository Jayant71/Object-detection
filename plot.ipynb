```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create some sample data
train_loss = {'box': [2.2, 2.0, 1.8, 1.6, 1.4, 1.2, 1.0],
              'cls': [2.5, 2.0, 1.5, 1.0, 0.8, 0.6, 0.5],
              'dfl': [1.3, 1.2, 1.1, 1.0, 0.9, 0.8, 0.6]}

val_loss = {'box': [1.8, 1.6, 1.4, 1.2, 1.0, 0.8, 0.6],
              'cls': [1.6, 1.4, 1.2, 1.0, 0.8, 0.6, 0.5],
              'dfl': [1.20, 1.15, 1.10, 1.05, 1.00, 0.95, 0.90]}

# Create a dictionary to store the precision and recall data
metrics = {'precision': [0.9, 0.8, 0.7, 0.6, 0.5],
            'recall': [0.9, 0.8, 0.7, 0.6, 0.5]}

# Create the line chart for training loss
sns.relplot(x=range(len(train_loss['box'])), y='box', data=train_loss, label='Box Loss')
sns.relplot(x=range(len(train_loss['box'])), y='cls', data=train_loss, label='Cls Loss')
sns.relplot(x=range(len(train_loss['box'])), y='dfl', data=train_loss, label='Dfl Loss')
plt.title('Training Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Create the line chart for validation loss
sns.relplot(x=range(len(val_loss['box'])), y='box', data=val_loss, label='Box Loss')
sns.relplot(x=range(len(val_loss['box'])), y='cls', data=val_loss, label='Cls Loss')
sns.relplot(x=range(len(val_loss['box'])), y='dfl', data=val_loss, label='Dfl Loss')
plt.title('Validation Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Create the PR curve
sns.lineplot(x=range(len(metrics['precision'])), y='precision', data=metrics, label='Precision')
sns.lineplot(x=range(len(metrics['precision'])), y='recall', data=metrics, label='Recall')
plt.title('Precision-Recall Curve')
plt.xlabel('Epoch')
plt.ylabel('Precision/Recall')
plt.legend()
plt.show()
```
